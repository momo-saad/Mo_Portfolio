# Data Loading and Initial Exploration

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset into a DataFrame
df = pd.read_excel(r"C:\Users\DELL\OneDrive\Desktop\projects\Python Projects\Project 1\order dataset.xlsx")
```

```python
# Display the first few rows to understand the data
df.head()
```

```python
# Display basic information about the DataFrame (column names, data types, etc.)
df.info()
```

```python
# Describe the dataset to get basic statistics
df.describe()
```

# Data Cleaning

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Remove rows with significant missing values or fill with appropriate values
df.dropna(inplace=True)
```

```python
print(df.isnull().sum())
```

```python
# Check for duplicates and remove them
df.drop_duplicates(inplace=True)
```

```python
# Convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Convert numeric columns to appropriate types
numeric_column = ['Order Quantity']
df[numeric_column] = df[numeric_column].astype(float)
```

```python
df.info()
```

# Exploratory Data Analysis (EDA)

```python
# Plot distribution of 'Order Year' to understand the time span of the data
sns.histplot(df['Order Year'], kde=True)
plt.title("Distribution of Order Year")
plt.show()
```

```python
# Visualize sales trends over time
df.groupby('Order Year')['Order Total'].sum().plot(kind='line')
plt.title("Total Sales Over the Years")
plt.xlabel("Year")
plt.ylabel("Total Sales $")
plt.show()
```

```python
# Analyze sales by customer type
sns.barplot(x='Customer Type', y='Order Total', data=df)
plt.title("Sales by Customer Type")
plt.show()
```

```python
# Analyze order quantity by product category
sns.boxplot(x='Product Category', y='Order Quantity', data=df)
plt.title("Order Quantity by Product Category")
plt.show()
```

# Data Analysis

```python
# Correlation analysis
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# Identify top-selling products
top_products = df.groupby('Product Name')['Order Total'].sum().sort_values(ascending=False).head(5)
top_products.plot(kind='barh')
plt.title("Top 5-Selling Products")
plt.show()
```

```python
# Analyze profit margins
sns.histplot(df['Profit Margin'], kde=True)
plt.title("Distribution of Profit Margins")
plt.show()
```

```python
# Analyze shipping costs by state
sns.barplot(x='State', y='Shipping Cost', data=df)
plt.title("Shipping Cost by State")
plt.xticks(rotation=90)
plt.show()
```
